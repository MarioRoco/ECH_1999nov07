```python
def create_spectroheliogram_from_interpolation(spectralimage_interpolated_list, w_px_range, w_px_range_bckg, slope_list, intercept_list):
    c1, c2 = w_px_range
    cb1, cb2 = w_px_range_bckg
    
    spectroheliogram_line_NT, spectroheliogram_line_nobg_NT  = [],[]
    for spectrum_array in spectralimage_interpolated_list:
    
        integrated_intensity_line_nobg_1col = []
        for row_ in range(len(slope_list)): 
            
            # Calculate the average of the profile of each row
            bckg_average = np.mean(spectrum_array[row_, cb1:cb2+1])
            integrated_intensity_line_nobg_1row1col = slope_list[row_] * (np.sum(spectrum_array[row_, c1:c2+1] - bckg_average) ) # Line without (background subtracted)
            
            # Save in lists
            integrated_intensity_line_nobg_1col.append(integrated_intensity_line_nobg_1row1col) # Line without background (background subtracted)
        
        # save the above lists (which represent the columns) in lists to crate the 2D-array of the spectroheliogram
        spectroheliogram_line_nobg_NT.append(integrated_intensity_line_nobg_1col) # Line without background (background subtracted)
        
    # Convert to array and transpose
    spectroheliogram_line_nobg = np.array(spectroheliogram_line_nobg_NT).T # Line without background (background subtracted)
    
    return spectroheliogram_line_nobg
```

```python
intensity_map = create_spectroheliogram_from_interpolation(spectralimage_interpolated_list=spectral_image_interpolated_list, w_px_range=w_px_range, w_px_range_bckg=w_px_range_bckg, slope_list=pixelscale_list, intercept_list=pixelscale_intercept_list)
```

```python
N_lat, N_lon = intensity_map.shape
x_edges = np.arange(0-0.5, N_lon-1+0.5)   # length Nx+1
y_edges = np.arange(0-0.5, N_lat-1+0.5)[::-1]  # length Ny+1

vmin_sumer, vmax_sumer = 8e-3, 1e-0


# Show intensity map
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(13, 6))
img = ax.imshow(intensity_map, cmap='Greys_r', norm=LogNorm(vmin=vmin_sumer, vmax=vmax_sumer), aspect='auto')
cbar = fig.colorbar(img, ax=ax, pad=0.03)
cbar.set_label(f'Integrated spectral radiance [W/sr/m^2]', fontsize=16)
ax.set_title(f'SOHO/SUMER intensity map ({round(wavelength_range_intensity_map[0],2)} - {round(wavelength_range_intensity_map[1],2)}) \u212B. Bckg subtracted.', fontsize=18)
ax.set_xlabel('Helioprojective longitude (pixels)', fontsize=16)
ax.set_ylabel('Helioprojective latitude (pixels)', fontsize=16)
ax.axis('auto') # Ensures equal scaling of axis x and y
plt.show(block=False)
```
